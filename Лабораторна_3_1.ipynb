**Ось приклад коду на пайтоні який виконує прості арифметичні дії**

```python
# Додавання
def addition(a, b):
    result = a + b
    return result

# Віднімання
def subtraction(a, b):
    result = a - b
    return result

# Множення
def multiplication(a, b):
    result = a * b
    return result

# Ділення
def division(a, b):
    if b == 0:
        return "Error: division by zero"
    result = a / b
    return result

# Приклад використання:
num1 = 10
num2 = 5

sum_result = addition(num1, num2)
diff_result = subtraction(num1, num2)
product_result = multiplication(num1, num2)
quotient_result = division(num1, num2)

print(f"Sum: {sum_result}")
print(f"Difference: {diff_result}")
print(f"Product: {product_result}")
print(f"Quotient: {quotient_result}")
```

*Зараз активно розвивається ШІ який пишеться більшості на пайтоні, тому саме час почати вивчати цю мову програмування*

*Ось міні код який аналізує погоду та температуру і рекомендує оптимальну дію на основі простих правил. Можна було  б розвивати цей приклад, додаючи більше умов та правил, щоб зробити його більш розумним.*

```python
# Простий штучний інтелект для визначення оптимальної дії на основі правил

# Ввід інформації
weather = "sunny"
temperature = 25

# Правила для вибору дії
if weather == "sunny" and temperature > 20:
    decision = "Go to the beach"
elif weather == "sunny":
    decision = "Go for a walk"
else:
    decision = "Stay at home"

# Виведення результату
print("Recommended action: " + decision)
```

*Цей код лише демонструє базовий принцип штучного інтелекту і не представляє собою складних інтелектуальних систем. Розробка справжніх інтелектуальних агентів вимагає глибокого розуміння та дослідження в областях машинного навчання, глибокого навчання, обробки природної мови, комп'ютерного бачення та інших.*
